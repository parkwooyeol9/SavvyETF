```python
# 수익률 데이터를 저장할 빈 데이터프레임 생성
etf_data = pd.DataFrame(columns=['Ticker', '1mo Return', '12mo Return'])
for ticker in etf_tickers:
    try:
        etf = yf.Ticker(ticker)
        periods = ['1mo', '1y']
        returns = {'Ticker': ticker}

        for period in periods:
            hist = etf.history(period=period)

            if not hist.empty and len(hist) > 1:  # Ensure we have at least two data points
                first_close = hist.iloc[0]['Close']
                last_close = hist.iloc[-1]['Close']
                return_rate = (last_close - first_close) / first_close
                if period == '1y':
                    returns['12mo Return'] = return_rate
                else:
                    returns[f'{period} Return'] = return_rate
            else:
                # Handle case with insufficient data
                if period == '1y':
                    returns['12mo Return'] = 0
                else:
                    returns[f'{period} Return'] = 0
                print(f"Warning: Insufficient data for {ticker} over {period} period")

        # 데이터프레임에 티커와 계산된 수익률 추가
        returns_df = pd.DataFrame([returns])
        # Avoid the FutureWarning by ensuring non-empty DataFrame before concatenation
        if not returns_df.empty:
            etf_data = pd.concat([etf_data, returns_df], ignore_index=True)
    except Exception as e:
        print(f"Error processing {ticker}: {e}")

# 결측치 처리
etf_data = etf_data.fillna(0)

# 1개월 수익률 기준 상위 5개 ETF 출력
print("Top 5 ETFs for 1mo Return:")
top_etfs = etf_data.sort_values(by='1mo Return', ascending=False).head(5)
print(top_etfs[['Ticker', '1mo Return']], '\n')

# 1개월 수익률 기준 하위 5개 ETF 출력
print("Bottom 5 ETFs for 1mo Return:")
bottom_etfs = etf_data.sort_values(by='1mo Return', ascending=True).head(5)
print(bottom_etfs[['Ticker', '1mo Return']], '\n')

# 12개월 수익률 정보도 저장
print("12-month returns for reference:")
print(etf_data[['Ticker', '12mo Return']].sort_values(by='12mo Return', ascending=False).head(5), '\n')

# Save the results to a CSV file for reference
etf_data.to_csv('etf_returns.csv', index=False)
print("Analysis complete. Results saved to 'etf_returns.csv'")

```
